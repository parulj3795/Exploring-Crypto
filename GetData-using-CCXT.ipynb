```python
import ccxt
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget
```

### Initialise Exchange

```python
exchange = ccxt.binance({
    'rateLimit': 1200,
    'enableRateLimit': True,
})
```

List of cryptocurrency exchanges available in CCXT.

```python
exchange = ccxt.binance()  # Replace 'binance' with your exchange of choice
markets = exchange.load_markets()
for symbol in exchange.symbols:
    print(symbol)
```

### Fetch Historical Data

```python
symbol = 'ETH/USDT'
timeframe = '1d'  # 1-day candles
since = exchange.parse8601('2022-01-01T00:00:00Z')  # Fetch data since January 1, 2022
limit = 365  # Number of data points to fetch

ohlcv = exchange.fetch_ohlcv(symbol, timeframe, since, limit)
```

```python
df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')  # Convert timestamps to datetime objects
df.set_index('timestamp', inplace=True)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['close'], label='Close Price')
plt.title(f'{symbol} Close Price')
plt.xlabel('Date')
plt.ylabel('Price (USDT)')
plt.legend()
plt.show()
```

### Popular Cryptocurrencies

1. Bitcoin (BTC): The first and most well-known cryptocurrency.
2. Ethereum (ETH): A decentralized platform that enables smart contracts and decentralized applications (DApps).
3. Ripple (XRP): Known for its digital payment protocol more than its cryptocurrency.
4. Litecoin (LTC): Created as the "silver to Bitcoin's gold", known for its faster transaction confirmation times.
5. Cardano (ADA): A project that aims to develop a more balanced and sustainable ecosystem for cryptocurrencies.
6. Polkadot (DOT): Aims to enable different blockchains to transfer messages and value in a trust-free fashion; seeking to make a web where users are in control.
7. Chainlink (LINK): A decentralized oracle network that aims to connect smart contracts with data from the real world.
8. Binance Coin (BNB): Originally created as a utility token for the Binance exchange, now used for various purposes within its ecosystem.
9. Solana (SOL): Known for its high performance and low transaction costs, aiming to provide fast, secure, scalable decentralized apps and marketplaces.
10. Dogecoin (DOGE): Started as a joke but has gained popularity and is used for tipping and donations.
